## cleaning the dataset

```python
import numpy as np
import pandas as pd
```

```python
RAW_DATA_PATH = r"/home/moris/Projects/expense/data/Monefy.Data.6-16-26.csv"
```

```python
df = pd.read_csv(RAW_DATA_PATH, usecols=['date', 'account', 'category', 'amount', 'description'])

df[100:150]
```

```python
# Remove negative sign and commas from amount column
df['amount'] = df['amount'].astype(str).str.replace('-', '', regex=False)
df['amount'] = df['amount'].astype(str).str.replace(',', '', regex=False).astype(float)

# Display a sample to verify the cleaning
print("Amount column after cleaning:")
print(df[['amount']].head(10))
```
